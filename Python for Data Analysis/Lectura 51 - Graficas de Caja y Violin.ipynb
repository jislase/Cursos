```python
import numpy as np
from numpy.random import randn
import pandas as pd

from scipy import stats

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
#Graficas de caja y de violin
url = 'http://en.wikipedia.org/wiki/Box_plot#mediaviewer/File:Boxplot_vs_PDF.svg'

#Creeamos dos distribuciones
data1 = randn(100)
data2 = randn(100) + 2 #Mover la media

#Creamos el grafico de caja
sns.boxplot(data=[data1,data2])
```

```python
#Podemos incluir los valores atipicos
sns.boxplot(data=[data1,data2],whis=np.inf)
```

```python
#Podemos mostrarlo de manera horizontal
sns.boxplot(data=[data1,data2],whis=np.inf, orient = "h")
```

```python
#Los graficos de violin pueden combinar la simplicidad de el grafico de caja con la informacion d eun KDE
```

```python
#Creamos un ejemplo donde el grafico de caja no nos da la imagen completa

#Distribucion Normal
data1 = stats.norm(0,5).rvs(100)

#Dos distribuciones gamma concatenadas (La segunda esta invertida)
data2 = np.concatenate([stats.gamma(5).rvs(50)-1,
                        -1*stats.gamma(5).rvs(50)])

#Graficamos
sns.boxplot(data=[data1,data2],whis=np.inf)
```

```python
#Con las graficas anteriores podriamos pensar que las distribuciones son similares
sns.violinplot(data=[data1,data2])
```

```python
#Podemos ver la diferencia de las distribuciones
```

```python
#Podemos ajustar el ancho de banda del kernel
sns.violinplot(data2,bw=0.01)
```

```python
#Parecido a una grafica de alfombra, Podemos incluir puntos individuales
sns.violinplot(data1,inner="stick")
```
