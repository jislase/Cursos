```python
import numpy as np
from numpy.random import randn
import pandas as pd

from scipy import stats

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
#Creamos un conjunto de datos
dataset = randn(100)

#Usamos distplot para combinar graficas, Por defecto pondra el kde sobre el histograma
sns.distplot(dataset,bins=25)
```

```python
#Se pueden modificar los paramentros
sns.distplot(dataset,rug=True,hist=False)
```

```python
#Para controlar las distintas graficas de distplot, se usa el argumento [grafica]_kws con diccionarios.
sns.distplot(dataset,bins=25,
             kde_kws={'color':'indianred','label':'KDE PLOT'},
             hist_kws={'color':'blue','label':"HISTOGRAM"})
```

```python
from pandas import Series

#Crear una serie de nuestro conjunto de datos
ser1 = Series(dataset,name='My_DATA')

#Graficar la Serie
sns.distplot(ser1,bins=25)
```
