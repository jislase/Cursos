```python
import numpy as np
from numpy.random import randn
import pandas as pd

from scipy import stats

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

#Comando para que aparezcan las graficas dentro de jupyter
%matplotlib inline
```

```python
#Creamos un conjunto de datos distribuido de forma normal
dataset1 = randn(100)

#Graficamos el conjunto de datos, el rango es de 10 por defecto
plt.hist(dataset1)
```

```python
#Hagamos otro conjunto de datos
dataset2 = randn(80)

#Graficamos
plt.hist(dataset2,color='indianred')
```

```python
#Podemos usar normed para graficar en el mismo grafico

#Ponemos normed=True Las graficas seran normalizadas para poder comparar los conjuntos de datos con distintos numeros de observaciones
#Ponemos alpha=0.5 Para la transparencia

plt.hist(dataset1,normed=True,color='indianred',alpha=0.5,bins=20)
plt.hist(dataset2,normed=True,alpha=0.5,bins=20)
```

```python
#Hacemos dos conjuntos de datos aleatorios con distribucion normal
data1 = randn(1000)
data2 = randn(1000)

#Podemos representar la union de las distribuciones usando joint plots
sns.jointplot(data1,data2)
```

```python
#Podemos usar rango Hexagonal para obtener una imagen mas precisa
sns.jointplot(data1,data2,kind='hex')
```
